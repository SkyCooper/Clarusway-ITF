### Code Challenge-3

With your \$ 200,  How many pieces of material can you get for $ 11 each? How much money do you have left after buying?

````python
money = 200
price = 11
piece_of_met = money // price
change = money % price
print(piece_of_met, 'piece')
print(change, 'dolar')
````
