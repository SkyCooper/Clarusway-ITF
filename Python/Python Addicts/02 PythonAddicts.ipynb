```python
# F3606 - Bahadır 

print("""**********************
KULLANICI GİRİŞ PANELİ
**********************""" )
kullanıcı_hakkı = 3
id1 = "aasuman"
pass1 = "012345"
while True:

    id = input("Kullanıcı Adı:")
    pas = input("Parola:")

    if (id == id1 and pas != pass1):
        print("Parolayı yeniden giriniz")
        kullanıcı_hakkı-=1
        print("Kalan giriş hakkı: {}".format(kullanıcı_hakkı))
    elif (id != id1 and pas == pass1):
        print("Kullanıcı adınız hatalı. Lütfen tekrar giriniz")
        kullanıcı_hakkı-=1
        print("Kalan giriş hakkı: {}".format(kullanıcı_hakkı))
    elif (id != id1 and pas != pass1):
        print("Kullanıcı adı ve parola hatalı")
        kullanıcı_hakkı-=1
        print("Kalan giriş hakkı: {}".format(kullanıcı_hakkı))
    else:
        print("Giriş başarılı.Hoşgeldiniz...")
        break
    if kullanıcı_hakkı == 0:
        print("Giriş hakkınız bitti.")
        break
```
