```python
"""
  Aşağıdaki ürünler tablosunda toptan satış fiyatları ve birimleri yer almaktadır.
  Belirtilen satış fiyatı 1 KG içindir ve her ürün birim miktarı (ve katları) kadar satılmaktadır.
  * Örnek: Elmanın toptan alışı "1KG = 10 TL'dir" ve "10 KG ve katları" kadar alınabilir.
  
    Ürün      Toptan    Birim
    -------------------------
    Elma      TL 10     KG 10
    Armut     TL 12     KG 5 
    Domates   TL 8      KG 20
    Biber     TL 20     KG 4 
"""
"""
  Soru-1: (Yukarıdaki tabloya göre)
  * Toptan ürün satın almak isteyen bir esnafın, sermayesini ve hangi üründen kaç kilo almak istediğini isteyiniz.
  * Sonuç olarak: Kaç KG alabileceğini ve para üstü miktarını yazdırınız.
  * Kurallar:
  * * Daha pahalı olan ürün, önce satılacaktır. Örnek: 20 KG biber ve 20 KG elma için önce biber satılacak, kalan paraya göre elma satılacaktır.
  ÖRNEK:
    Sermaye giriniz (TL): 1200
    Elma (KG): 25
    Armut (KG): 25
    Domates (KG): 25
    Biber (KG): 25
    Sonuc: "Biber: 24 KG + Armut: 25 KG + Elma: 20 KG + Domates: 20 KG = 1140 TL * Para: 1200 TL, ParaUstu: 60 TL"
"""

# Toptan Ürün Bilgileri = dict():
toptan = {
    "Elma": {"fiyat": 10, "birim": 10},
    "Armut": {"fiyat": 12, "birim": 5},
    "Domates": {"fiyat": 8, "birim": 20},
    "Biber": {"fiyat": 20, "birim": 4}
}

# Kullanıcı Seçimleri = dict():
secim = {
    "sermaye": input("Lütfen, para miktarı giriniz (TL): "), # Sermaye bilgisi iste
    "miktar": {}
}
for x in toptan:
  secim["miktar"][x] = int(input(x + ' (KG): ')) # KG bilgisi iste

# Ürünleri fiyatına göre sırala (pahalıdan-ucuza):
urunler = dict(sorted(toptan.items(), key = lambda x: x[1]["fiyat"], reverse = 1))

# İstenen miktar ve birim miktara göre, miktar ve tutar hesapla:
toplam_miktar = {} # Çıktıda miktar=0 olanları da göstermek için => dict(map(lambda x: (x, 0), urunler.keys()))
toplam_tutar = 0
kalan_tutar = int(secim["sermaye"])
for urun, detay in urunler.items():
  # Hesapla:
  urun_miktar = detay["birim"]*(secim["miktar"][urun]//detay["birim"]) # Alınabilecek max. miktar.
  urun_tutar = detay["fiyat"]*urun_miktar # Miktarın tutarı.
  if (kalan_tutar >= urun_tutar): # Yeteri kadar KalanTutar var, o zaman satın al.
    # Sonuca Yaz:
    toplam_miktar[urun] = urun_miktar
    toplam_tutar += urun_tutar
    # KalanTutar:
    kalan_tutar -= urun_tutar
  elif (kalan_tutar >= (detay["birim"]*detay["fiyat"])): # KalanTutar var ama yeteri kadar yok. Ama birim miktar kadar ürün alabilir.
    # Hesapla:
    urun_miktar = int(kalan_tutar/detay["fiyat"]) # KalanTutar ile kaç kilo alınabilir.
    urun_miktar = detay["birim"]*(urun_miktar//detay["birim"]) # Alınabilecek max. miktar.
    urun_tutar = detay["fiyat"]*urun_miktar # Miktarın tutarı.
    # Sonuca Yaz:
    toplam_miktar[urun] = urun_miktar
    toplam_tutar += urun_tutar
    # KalanTutar:
    kalan_tutar -= urun_tutar
  else: # KalanTutar ürün satın almak için yetersiz.
    break # Geri kalan ürünlere bakmaya gerek yok.

# Yazdırma için çıktı oluştur:
cikti = ''
for urun, miktar in toplam_miktar.items():
  cikti += urun + ': ' + str(miktar) + ' KG + '
cikti = cikti.rstrip(' + ') # Sağdaki fazla yazıyı sil.
cikti += ' = ' + str(toplam_tutar) + ' TL * Para: ' + secim["sermaye"] + ' TL, ParaUstu: ' + str(kalan_tutar) + ' TL'

# Çıktıyı yazdır:
print(cikti)
```

```python
"""
  Aşağıdaki ürünler tablosunda toptan satış fiyatları ve birimleri yer almaktadır.
  Belirtilen satış fiyatı 1 KG içindir ve her ürün birim miktarı (ve katları) kadar satılmaktadır.
  * Örnek: Elmanın toptan alışı "1KG = 10 TL'dir" ve "10 KG ve katları" kadar alınabilir.
  
    Ürün      Toptan    Birim
    -------------------------
    Elma      TL 10     KG 10
    Armut     TL 12     KG 5 
    Domates   TL 8      KG 20
    Biber     TL 20     KG 4 
"""
"""
  Soru-2: (Yukarıdaki tabloya göre)
  * Pazarda ürün satan bir esnafın elindeki ürün miktarını ve satış fiyatını ve tahmini fire miktarını isteyiniz.
  * Sonuç olarak: Alış, Satış, Fire ve Kar/Zarar bilgilerini hesaplayınız.
  * Kurallar:
  * * Fire miktarı bir defa istenecektir. (Tüm ürünlere aynı fire oranı hesaplanacaktır.)
  * * Fire ürünlere ayrı-ayrı uyglanacaktır. (Toplam miktar üzerinden yapılmayacaktır.)
  * * Satış birimi 1 KG'dir. Yani fire hesabından çıkan küsüratlarda fire sayılacaktır. Örneğin: "25 KG" - "%10 Fire" = 22.5 değil "22 KG" sayılacaktır.
  ÖRNEK:
    Elma (KG * TL): 20 * 18
    Armut (KG * TL): 25 * 20
    Domates (KG * TL): 20 * 12
    Biber (KG * TL): 24 * 40
    Fire (%): 10
    Sonuc: "Alış-Miktar: 89 KG, Alış-Tutar: 1140 TL * Satış-Miktar: 79 KG, Satış-Tutar: 1820 TL * Fire-Miktar: 10 KG, Fire-Tutar: 132 TL * SONUÇ: 680 TL KAR"
"""

# Toptan Ürün Bilgileri = dict():
toptan = {
    "Elma": {"fiyat": 10, "birim": 10},
    "Armut": {"fiyat": 12, "birim": 5},
    "Domates": {"fiyat": 8, "birim": 20},
    "Biber": {"fiyat": 20, "birim": 4}
}

# Kullanıcı Seçimleri = dict():
secim = {
    "urun": {}
}
for x in toptan:
  secim["urun"][x] = input(x + ' (KG * TL): ').replace('*', ' ') # KG*TL bilgisi iste. (önce KG * Sonra TL)
  secim["urun"][x] = tuple(map(lambda x: int(x.strip()), secim["urun"][x].split())) # Parçala ve boşlukları temizle ve sayıya çevir.
# Fire bilgisini de iste:
secim["fire"] = int(input('Fire (%): '))

# Sonuç verisi:
sonuc = {
    "alis": {"miktar": 0, "tutar": 0},
    "satis": {"miktar": 0, "tutar": 0},
    "fire": {"miktar": 0, "tutar": 0}
}
# Sonuç hesapla ve işle:
for urun, detay in secim["urun"].items(): # detay[0] = KG, detay[1] = TL
  # Alış Verisi Ekle:
  sonuc["alis"]["miktar"] += detay[0]
  sonuc["alis"]["tutar"] += detay[0]*toptan[urun]["fiyat"]
  # Satış ve Fire Hesapla:
  satis_miktar = int(detay[0]*(1-(secim["fire"]/100))) # (miktar * 0.Fire) ile hesapla ve küsüratı sil.
  fire_miktar = detay[0] - satis_miktar
  # Satış Verisi Ekle:
  sonuc["satis"]["miktar"] += satis_miktar
  sonuc["satis"]["tutar"] += satis_miktar*detay[1]
  # Fire Verisi Ekle:
  sonuc["fire"]["miktar"] += fire_miktar
  sonuc["fire"]["tutar"] += fire_miktar*toptan[urun]["fiyat"]

# Kar-Zarar hesapla:
  fark = sonuc["satis"]["tutar"]-sonuc["alis"]["tutar"]
  durum = ('KAR' if fark>= 0 else 'ZARAR')

# Sonucu Yazdır:
print(
  f'Alış-Miktar: {sonuc["alis"]["miktar"]} KG, Alış-Tutar: {sonuc["alis"]["tutar"]} TL *',
  f'Satış-Miktar: {sonuc["satis"]["miktar"]} KG, Satış-Tutar: {sonuc["satis"]["tutar"]} TL *',
  f'Fire-Miktar: {sonuc["fire"]["miktar"]} KG, Fire-Tutar: {sonuc["fire"]["tutar"]} TL *',
  f'SONUÇ: {fark} TL {durum}'
)
```
