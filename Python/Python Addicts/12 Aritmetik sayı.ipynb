```python
num = int(input("Enter a number>>>: "))
divisor = []
for i in range(1,num+1):
    if num % i == 0:
        divisor.append(i)
if sum(divisor) % len(divisor)==0:
    print(f"Yes {num} is Arithmetic")
else:
    print(f"Yes {num} is not Arithmetic")
```
