```python
# F3785 - Abdullah

# 40 saat altı saatlik ücret tarifesi = 100 $
# 40 saat üstü saatlik ücret tarifesi = 150 $
çalışma_saati = int(input('Haftada kaç saat çalışıyorsunuz: '))
ücret = 0
if çalışma_saati < 40 :
  m = çalışma_saati * 100
  ücret += m
elif çalışma_saati > 40 :
  n = (çalışma_saati-40) * 150
  ücret += n
print(f"Haftalık Ücretiniz: {ücret} $ dır.")
```
