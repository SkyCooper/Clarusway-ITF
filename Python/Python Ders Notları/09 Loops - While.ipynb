As one of the most main functions in programming, loops are an important part of almost every programming language. Loops enable programmers to set certain sections of their code to repeat through a number of loops which are referred to as iterations.

```python
answer = 44

question = 'What number am I thinking of?  '
print ("Let's play the guessing game!")

while True:
    guess = int(input(question))

    if guess < answer:
        print('Little higher')
    elif guess > answer:
        print('Little lower')
    else:  # guess == answer
        print('Are you a MINDREADER!!!')
        break
```

```python
sentence = input("Enter a sentence : ")
word_list = sentence.split()


longest = 0 # her seferinde uzun kelimeyi bulunca burasÄ± deÄiÅecek
i = 0 # index numarsÄ±nÄ± tekrar ettirmek iÃ§in atandÄ±, her seferinde artacak

while i < len(word_list) : # cÃ¼mlenin uzunluÄu 5, yani iÃ§inde 0,1,2,3,4 e kadar index var. i=4 oluncaya kadar Ã§alÄ±Å
    if len(word_list[i]) > longest :                                                     # i=5 olursa False olur dÃ¶ngÃ¼ biter.
        longest = len(word_list[i])
    i += 1 # counter'Ä± if bloÄu dÄ±ÅÄ±na alÄ±yoruz. ÃNEMLÄ°

print("The lenght of the longest word :", longest)
```

```python
# Write a program that takes a number from the user, assigns it to the variable (number),
# and prints the squares of all numbers between 0 and the number (not included) entered by the user.
# Note: Write the program with while statement.

number = int(input('Please enter a number: '))
counter = 0
while counter < number:
    print(counter**2)
    counter += 1
```
