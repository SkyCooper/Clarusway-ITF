# enumerate

```python
harfler = ["a", "b", "c", "d", "e"]
for harf in harfler:
    print(harf) # harfleri sırayla yazdırdı.
```

```python
harfler = ["a", "b", "c", "d", "e"]
for harf in enumerate(harfler):
    print(harf) # harfleri sırayla ve index numaralarıyla yazdırdı.
                # ama default değeri sıfır olduğu için böyle yaptı.
```

```python
# ikili çıktı ürettiği için sadece numaraları almak istersem;
harfler = ["a", "b", "c", "d", "e"]
for index, harf in enumerate(harfler):
    print(index) 
```

```python
# ikili çıktı ürettiği için sadece harfleri almak istersem;
harfler = ["a", "b", "c", "d", "e"]
for index, harf in enumerate(harfler):
    print(harf) 
```

```python
# ikili çıktı ürettiği için hem harfleri / hem numaraları almak istersem;
harfler = ["a", "b", "c", "d", "e"]
for index, harf in enumerate(harfler):
    print(index, harf) 
```

```python
# daha görsel bir çıktı için;

for index, harf in enumerate(harfler):
    print("{}. harf : {}". format(index+1, harf)) 
```

```python
# aynı çıktıyı saydırmayı 1'den başlatarak da yapabilirim.

for index, harf in enumerate(harfler, 1):
    print("{}. harf : {}". format(index, harf))
```

# zip --> iterable objectlere uygulanır.

```python
# In some cases, you will need to set up the for loop with multiple variables and the iterables.
# Examine the example :

text = ['one','two','three','four','five']
numbers = [1, 2, 3, 4, 5]
for i in zip(text, numbers):
    print(i)  # -----> enumerate gibi eşleştirme listesi verir. Ama ikili, üçlü, dörtlü vs. olabilir.
```

```python
text = ['one','two','three','four','five']
numbers = [1, 2, 3, 4, 5]
for x, y in zip(text, numbers):
    print(x, ':', y) # --->ikili olduğu için ayrı ayrı da erişim yapabiliriz.
```

```python
for x, y in zip(text, numbers):
    print(x) # ----> veya sadece birisinin çıktısına ulaşabiliriz.
```

```python
# zip() fermuar gibi karşılıklı elemaları eşleştiriyor. çıktısı TUPLE olur.
zipped = zip(text, numbers) # aynı range gibi (on-the-air) object yaratır, çıktı vermez.
zipped
```

```python
print(* zipped) # elemanlar tükenene kadar eşleştirme yapar.               
```

```python
list(zipped)  # iterate edildi, yani eşleştirdi. onun için içi boşaldı
```

```python
_  # ---> içinde en son aldığı elemanın değerini tutar.
```

```python
x, y, *z = (11, 22, 33, 44, 55) # ilk iki değer x/y diğer üç değer z
print(x, y, z)
```

```python
x, y, *_ = (11, 22, 33, 44, 55) # ilk iki değer x/y diğer üç değer havada, atama olmadı
print(x, y)
```

```python
x, y, *z, t = (11, 22, 33, 44, 55, 66, 77) # asteriks-->çoklama demektir.
# x birinci, y ikinci, t en son kalan, z geriye kalanlar
print(x, y, z, t)
```
