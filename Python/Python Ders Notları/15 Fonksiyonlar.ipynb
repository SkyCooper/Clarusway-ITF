```python
def absulate_value(num):
    """ Burası fonksiyonun docstringidir."""
    if num < 0 :
        return -num
    elif num > 0:
        return num
```

```python
print(absulate_value.__doc__) # fonksiyonun docsrtingini görmek için .__doc__
```

```python
absulate_value(-8)
```

```python
abs(-8) # built in fonksiyon ile yapımı
```

# Walrus Operatör Kullanımı 

```python
# kahoot sorusu;
b = 0
d = []
while (b:= b+1) < 4: # 0 olan b'yi b+1 e şitledi ve b=1 oldu, listeye ekledi
    d.append(b)      # 1 olan b'yi b+1 e şitledi ve b=2 oldu, listeye ekledi
print(d)             # 1 olan b'yi b+1 e şitledi ve b=3 oldu, listeye ekledi
                     # b < 4 yani en son 3<4' ten sonra döngü çalışmadı ve bitti.
```
