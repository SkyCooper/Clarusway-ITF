# .get() kullanÄ±mÄ±

```python
pairs = {1: "apple",
    "orange": [2, 3, 4], 
    True: False, 
    None: "True",
}

print(pairs.get("orange", "not in dictionary"))
print(pairs.get(7))
print(pairs.get(12345, "not in dictionary"))
```

```python
books = {
    "Life of Pi": "Adventure Fiction", 
    "The Three Musketeers": "Historical Adventure",
    "Watchmen": "Comics", 
    "Bird Box": "Horror",
    "Harry Potter":"Fantasy Fiction",
    "Good Omens": "Comedy"
}
```

```python
# if/ else ile sorgulama;

book = input()
if book in books:
    print(books[book])
else:
    print("Book not found")
```

```python
# .get() method --> tek satÄ±r
book = input()

print(books.get(book, "Book not found"))
```
