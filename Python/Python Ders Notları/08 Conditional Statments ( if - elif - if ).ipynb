# Structure of the 'if' Statements
In some cases, your program needs to execute some part of the code only if a specific condition is true. The if statements check the condition. The condition is always a Boolean expression, that is, its value equals either True or False.

If it evaluates to True, it executes the body of the if statement. If it evaluates to False, it skips the body. This logic works like the English language.

```python
# if condition: ----> KOŞUL true ise aşağıdaki KOD BODY çalışsın demek.
    # body  # the body code should start at indentation (four spaces)
    
# condition, her halukarda (boolen) sonuç verecek bir ifade olmalı
# True / False olabilir
# Truty / Falsie de olabilir  if 1: (Truty) gibi
```

# Comparison Operators
Boolean values basically make it clear if a piece of code needs to be executed.

Because comparisons result in bool, it's always best to use them as a condition.

```python
# !=  eşit değildir,         not equal
# ==  eşittir,               equal
# >   büyüktür,              greater than
# <   küçüktür,              less than
# >=  büyüktür eşittir,      greater than or equal
# <=  küçüktür eşittir,      less than or equal
```

```python
x = 6
y = 9
print ("is x equal to y?                 :" , x == y)
print ("is x not equal to y?             :" , x != y)
print ("is x less than y?                :" , x < y)
print ("is x greater than y?             :" , x > y)
print ("is x less than or equal to y?    :" , x <= y)
print ("is x greater than or equal to y? :" , x >= y)
```

#  if-elif-else Statements

In Python you can deﬁne a series of conditionals using :

if for the ﬁrst one,

elif for the rest, up until the ﬁnal (optional),

else for anything not caught by the other conditionals.

```python
#if condition1:
    #execute body1

#elif condition2:
    #execute body2

#else:
    #execute body3
```

```python
weight = 80

if weight > 100:
    print("That's too heavy!")
elif weight > 75:
    print("I can lift that!")
else:
    print("That's too light!")
```

```python
audience = "baby"

if audience == "kid":
    print("it is free to go to cinema")
elif audience == "teen":
    print("discounted price!")
elif audience == "adult":
    print("normal price")
else:
    print("No such audience, stay at your home!")
```

```python
num1 = float(input("Enter first number :"))
num2 = float(input("Enter second number :"))
num3 = float(input("Enter third number :"))
if (num1 >= num2) and (num1 >= num3):
    largest = num1
elif (num2 > num3) and (num2 >= num1):
    largest = num2
else:
    largest = num3
print("The largest number is", largest)
```

# Nested 'if-elif-else' Statements

```python
audience_group = 'kid', 'teen', 'adult' # parantez olmasa bile virgül ile ayrıldığı inin TUPLE dir.

audience = "teen"

if audience in audience_group:
    if audience == "kid":
        print("it is free to go to cinema")
    elif audience == "teen":
        print("discounted price!")
    else: # audience == "adult":
        print("normal price")
else:
    print("No such audience, stay at your home!")
```
