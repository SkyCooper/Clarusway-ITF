\n represents next line.

Any mathematics operator can be used before the = character to make an in-place operation:

-= decrements the variable in place,

+= increment the variable in place,

*= multiply the variable in place,

/= divide the variable in place,

//= ﬂoor divide the variable in place,

%= returns the modulus of the variable in place,

**= raise to power in place.

```python
x = 5
print ('value of x       : ', x)

x += 2
print ("2 more of x      : ", x, "\n") # using string expression '\n', 
                                       # we produce extra line. 
                                       # So that we had empty line.
y = 10
print ('value of y       : ', y)

y -= 2
print ("2 minus y        : ", y, "\n")

z = 6
print ('value of z       : ', z)

z *= 2
print ("2 times z        : ", z, "\n")
```

```python
# Remember. There is no difference between
# ' '         " "       """  """

fruit = 'Orange'
vegetable = "Tomato"
print (fruit, " and " , vegetable)

fruit = 'Orange'
vegetable = "Tomato"
print (fruit, ' and ' , vegetable)


fruit = 'Orange'
vegetable = "Tomato"
print (fruit, """ and """ , vegetable)
```

BASIC ESCAPE SEQUENCES  (STRİNG İÇİNDE)

1. ÖZEL KARAKTERLERİN SİHRİNİ YOK EDER KULLANILIR.
2. BAZI KARAKTERLERE SİHİR KATAR

\n : means new line,  YENİ SATIR

\t : means tab mark, TAB 4 BOŞLUK OLUYOR

\b : means backspace. It moves the cursor one character to the left. BİR KARAKTER SOLA ALIR.

```python
print('Ahmet'in ceketi') # hata verir
```

```python
print('Ahmet\'in ceketi') # tırnağın özel karakter özelliğini yok eder.
                          # NORMAL SIRADAN BİR KARAKTER HALİNE GETİRDİ
```

```python
print('Ahmet\in ceketi')  # ÖZEL BİR KARAKTERLE KULLANMADIĞI İÇİN NORMAL BİR KARAKTER OLDU
```

```python
print('Ahmet\nceketi') # \n karakteri ESCAPE SQUENCE olacak
                        # new line olarak gördü
```

```python
print('Ahmet\\n ceketi') # bir ters slash ekleyip normal karakter yaptık
```

```python
print('C:\\north pole\noise_penguins.txt')
print('---------')
print('bir', 'iki', 'üç')
print('bir', 'iki', 'üç', sep='\t')   # aralarına bi tab koydu.
print('bir', 'iki', 'üç', sep='\n')   # hepsini yeni saturda yazdı.
print('bir ', 'iki ', 'üç', sep='\b') # boşluları sildi
print('bir', 'iki', 'üç', sep='\b')   # boşluk olmadığı için bi(r) ve ik(i) silindi.
```

```python
print('we are',         '\boosting', 'our',     '\brotherhood') # stringler arasında uzuuuun boşluk olsa bile 
                                                                # yanındaki stringin son karakterini siler.
print()
print('we are', '\toosting', 'our', '\brotherhood')
print()
print('we are', '\noosting', 'our', '\trotherhood')
```

```python
print('C:\\november\number_expenditure.txt') # \n numbera birleşik yazılmış ama kod bunu ayırıcı olarak görür. 
print()
print('C:\\november\ number_expenditure.txt') # böyle olsa doğru yazar
```

```python
print("one", "two", "three", sep="\t") # separated by tab marks
print()
print("one", "two", "three", sep="\n") # böyle olursa diğer satıra geçer
print()
print("one", "two", "three")
print("one", "two", "three", sep=" \b") # \b'den önce boşluk olduğu için stringden silmedi.
print("one", "two", "three", sep="\b")  # \b'den önce boşluk olmadığı için önceki stringin son karakteri silindi.
```

```python
print('we', 'are', 'united') # normalde böyle görünürdü

print('we', '\bare', '\bunited') # remember, normally print() function
                                 # separates expressions by spaces
                                 # bu komut bir önceki karakteri sildiği için boşluklar kaybolur.

print('we', 'are', 'united', sep=' \b') # aynı çıktıyı verir.\b'den önce boşluk
print('we', 'are', 'united', sep='\b')  # \b'den önce boşluk olmazsa stringlerden karakter siler
```
