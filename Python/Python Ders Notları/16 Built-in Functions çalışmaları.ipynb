# all()¶
bir dizi içinde bulunan bütün değerler True ise True değeri, eğer bu değerlerden herhangi biri False ise de False değeri döndürmektir.

```python
liste = [1, 2, 3, 4]
all(liste)
```

```python
liste = ['ahmet', 'mehmet', ''] #Listede False değerine sahip bir boş karakter dizisi bulunduğu için
all(liste)
```

```python
a = 3
t1 = a == 3         #sayı 3 mü?
t2 = a < 4          #sayı 4'ten küçük mü?
t3 = a % 2 == 1     #sayı bir tek sayı mı?
all([t1, t2, t3])   #sayı bu özelliklerin hepsine sahip mi?
```

```python
all((1, 2, "")) # içine itarable alır, içine tuple koyduk
```

```python
all("") # sadece boş olursa true
```

# filter()
all() ve any() bunun içinde kullanılabilir

```python
filter(all, iterable) ---> içine bir fonksiyon ve iterable alır.
fonksiyon içinde foksiyon çağırırken ismini yazıyoruz, () kullanmıyoruz.
```

```python
# fonksiyonla filitreleme yapılması;
def cift_sayilari_filtrele(x):
    if x%2==0:
        return x
    else:
        return None
```

```python
cift_sayilari_filtrele(3)
```

```python
# aynı fonksiyonun daha pythonista yazılması;
def cift_sayilari_filtrele(x):
    return x if x%2==0 else None
```

```python
cift_sayilari_filtrele(4)
```

```python
# for döngüsü ile listeden çiftleri seçmek;
sayilar = [*range(1,10)]
output = []
for i in sayilar:
    if i%2==0:
        output.append(i)
print(output)
```

```python
# filter ile yapılması;
sayilar = [*range(1,10)]
filter(cift_sayilari_filtrele, sayilar) # object üretir.
```

```python
sayilar = [*range(1,10)]
list(filter(cift_sayilari_filtrele, sayilar)) #çiftleri ayırdı.
```

```python
lista = ["ali", "deli", False, 1, "0", "", "veli", 0]
```

```python
filter()
```

```python
# bir text'te sesli harfleri süzen bir fonksiyon yazalım.

cümle = "ben size ders anlatırken huzur buluyorum. dersleri unutuyorum."
def vowel(letter):
    vowels = "a e ı i o ö ü u".split()
    if letter.lower() in vowels:
        return True
    else:
        return False
print(list(filter(vowel, cümle)))
```

# map ()

```python
def karesi(n):
    return n ** 2
```

```python
karesi(5)
```

```python
sayilar = list(range(1,6))
sayilar
```

```python
# map kullanmadan bir listenin hepsinin karesini almak;
sayilar = list(range(1,6)) 
print(sayilar)
for i in range(len(sayilar)):
    sayilar[i] = karesi(sayilar[i]) # indexleyerek karesini aldı.
print(sayilar)
```

```python
# map kullanarak;
sayilar = list(range(1,6)) 
map(karesi, sayilar) # on the air map object üretir.
```

```python
sayilar = list(range(1,6))
print(sayilar)
[*map(karesi, sayilar)] # çıktısını görmek için
```

```python
listem = [1, 4, 5, 4, 2, 9, 10]
```

```python
list(map(karesi, listem))
```

# enumarate()

```python
# enumarte olmadan uzun hali

isimler = ["ali", "deli", "veli"]
sayaç = 0
liste = []
for isim in isimler:
    liste.append((sayaç, isim))
    sayaç +=1
print(liste)
```

```python
print(list(enumerate(isimler)))  # elemanlara index numarası verip tuple yapıyor.
```

```python
grocery = ["bread", "water", "olive"]
enum_grocery = enumerate(grocery)
print(type(enum_grocery))
print(list(enum_grocery))
enum_grocery = enumerate(grocery, 10)
print(list(enum_grocery))
```

```python
print(*enum_grocery)
```

```python
enum_grocery = enumerate(grocery)
```

```python
dict(enum_grocery)
```

```python
dict(enumerate(grocery))
```
