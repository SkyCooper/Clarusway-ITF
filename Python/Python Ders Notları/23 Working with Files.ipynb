# Reading the Files with Loops

```python
# the last method for reading a file is using the lines of the file-like object as an iterator. 
sea = open("fishes.txt", 'r')   

for line in sea:
    print(line)

sea.close()
```

```python
# Since the type of sea.readlines() object is a list, we can use it as an iterator and read the whole file in the same way

sea = open("fishes.txt", 'r')   

for line in sea.readlines():
    print(line)

sea.close()
```

```python
 # we have two most commonly used modes ('w' and 'a') that allow us to modify the files. Let's examine them one by one :
```

```python
# Let's create and write string data to a file. We're going to use .write() method for writing :

with open("dummy_file.txt", 'w', encoding="utf-8") as file:  
# rastgele isim verip bir dosya oluşturdum ve w modunda açtım.

    file.write('This is the first line of my text file')  
    # içine yazmaya başladım string olarak

with open("dummy_file.txt", 'r', encoding="utf-8") as file:
    print(file.read())  # reads the content of the 'dummy_file'
```

```python
# Now let's repeat the process and see what happens. This time the file (dummy_file) exists :

with open("dummy_file.txt", 'w', encoding="utf-8") as file:
    
    file.write('This is the new line for my dummy_file')  
    # yeni bir string yazılım

with open("dummy_file.txt", 'r', encoding="utf-8") as file:
    print(file.read())  # reads the content of the 'dummy_file'
    
# As you can see in the output, the previous content has been deleted and replaced with new content.
# overwrite --> üzerine yazdı.
```

```python
with open("dummy_file.txt", 'w', encoding="utf-8") as file:
    file.write('My first sentence')
    file.write('My second sentence,')
    file.write('My third sentence\n')
    file.write('My fourth sentence ')
    file.write('My last sentence')

with open("dummy_file.txt", 'r', encoding="utf-8") as file:
    print(file.read())
```

```python
# Now, think of that we have a list of fruit names. 
# Let's write them to a file each on separate lines one after another.

fruits = ['Banana', 'Orange', 'Apple', 'Strawberry', 'Cherry']

with open("fruits.txt", 'w', encoding="utf-8") as file: # yazılacağı için w modda açılır.
    for basket in fruits:
        file.write(basket + '\n')  # adds a newline character to each string
                                   # alt alta yazdırmak için
            
with open("fruits.txt", 'r', encoding="utf-8") as file: # okunacağı için r modda açılır.
    print(file.read())
```

# Writing to File with '.writelines' Method

```python
fruits = ['Banana\n', 'Orange\n', 'Apple\n', 'Strawberry\n', 'Cherry\n']

with open("fruits.txt", 'w', encoding="utf-8") as file:
    file.writelines(fruits)  # takes an iterator for writing
   
with open("fruits.txt", 'r', encoding="utf-8") as file:
    print(file.read())

with open("fruits.txt", 'r', encoding="utf-8") as file:
    print(file.readlines())
```

```python
with open('fishes.txt', 'r', encoding = "utf-8") as sea:
    fish_lines = sea.readlines()

with open('fishes.txt', 'w', encoding = "utf-8") as sea:
    for reef in fish_lines:
        sea.write(reef + '\n')
        
with open('fishes.txt', 'r', encoding = "utf-8") as sea:
    print(sea.read())
```

# Appending to File using 'a' Mode

```python
fruits = ['Banana\n', 'Orange\n', 'Apple\n', 'Strawberry\n', 'Cherry\n']

with open("fruits.txt", 'w', encoding="utf-8") as file:
    file.writelines(fruits)  # creates a file containing the elements of the list

with open("fruits.txt", 'r', encoding="utf-8") as file:
    print(file.read())  # reads the content of the file

with open("fruits.txt", 'a', encoding="utf-8") as file:
    file.write('Melon\n')  # adds "Melon" to the end of the existing file
   
with open("fruits.txt", 'r', encoding="utf-8") as file:
    print(file.read())  # reads the last content of the file

with open("fruits.txt", 'r', encoding="utf-8") as file:
    print(file.readlines())
```

```python
with open("istiklal.txt", 'r', encoding="utf-8") as file:
    print(file.read()) # bunu kıtalar arasında boluk olacak şekilde yazdır. 
```

```python
with open("istiklal.txt", 'r', encoding="utf-8") as file:
    lines = file.readlines() # istiklal marşı satırlarından oluşan listem oldu.
```

```python
lines # artık iterable olan bir listem var.
```

```python
counter = 0

with open("istiklal.txt", 'w', encoding="utf-8") as file: # üstüne yazmak için w modunda açılır.
    for i in lines:
        counter += 1
        
        if counter % 4 == 0:
            file.write(i + "\n") #4ncü satırı yaz ve newline yap
        else:
            file.write(i) # yoksa sadece satırı yaz
```

```python
with open("istiklal.txt", 'r', encoding="utf-8") as file:
    print(file.read())
```

```python
# tekrar boşlukları kaldırıp altına yazdırmak istiyorum 
with open("istiklal.txt", 'r', encoding="utf-8") as file:
    lines_new = file.readlines()
```

```python
lines_new
```

```python
with open("istiklal.txt", 'a', encoding="utf-8") as file: # ekleyerek yazmak için a modunda açılır.
    for i in lines_new:
        if i == "\n" : # boşluk varsa geç yazma
            pass
        else:
            file.write(i) # diğer satırları yaz
            
with open("istiklal.txt", 'r', encoding="utf-8") as file:
    print(file.read())
```
