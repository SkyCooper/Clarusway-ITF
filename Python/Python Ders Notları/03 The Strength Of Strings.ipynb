```python
sentence = "I live and work in Virginia"

print(sentence.upper()) # hepsi büyük
print(sentence.lower()) # hepsi küçük
print(sentence.swapcase()) # büyükler/küçük küçükler/büyük 
print(sentence)  # note that, source text is unchanged
print(sentence.title()) # kelimelerin ilk harfleri büyük
print(sentence.capitalize()) # strig'in ilk harfi büyük
changed_sentence = sentence.replace("i", "+") # i leri + yaptı
print(changed_sentence)
print(sentence.count('i'))  # kaç tane i harfi var onu sayar
```

```python
sentence = "I live and work in Virginia"

print(sentence.upper()) # hepsini büyük harfle yazdırır
print(sentence.lower()) # hepsini küçük harfle yazdırır
print(sentence.swapcase()) # büyük harfi küçük harfe çevirir ve bunun tersi de geçerlidir.
print(sentence.capitalize()) # dizenin ilk karakterini büyük harf yapar,
                             #ve geri kalanını küçük harfe değiştirir.
print(sentence.replace("i","*")) #belirtilen karakteri yenisiyle değiştirir
print(sentence.title()) # kelimelerin baş harfini büyük yazar 
print(sentence.strip("I,a")) # her iki taraftaki tüm boşlukları 
                             #(veya belirtilen karakterleri) kaldırır.
print(sentence.rstrip('nia')) # sağ taraftaki boşlukları (veya belirtilen karakterleri) kaldırır.
print(sentence.lstrip('I li')) # sol taraftaki boşlukları (veya belirtilen karakterleri) kaldırır.
print(sentence.count("i")) # dizedeki karaktein kaç defa tekrarlandığını gösteir.
```
