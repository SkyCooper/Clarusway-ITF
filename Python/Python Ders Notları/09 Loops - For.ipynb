```python
# liste yapÄ±p len() ile Ã§Ã¶zÃ¼mÃ¼

example_list = [11, 2, 24, 61, 48, 33, 3]
evens = []
odds = []

for i in example_list:
    if i % 2 == 0 :
        evens.append(i)
    else:
        odds.append(i)
print("The number of even numbers :", len(evens))
print("The number of odds numbers :", len(odds))
```

# Nested 'for' Loop

```python
# In Python, you can easily place one loop inside another one.
# First outer loop then inner one runs.
# You'll see it in the following example :
```

```python
who = ['I am ', 'You are ']
mood = ['happy', 'confident']
for i in who:
    for ii in mood:
        print(i + ii)
```

```python
names = ["cooper", "adil", "gregor"]
mood = ["happy", "sad"]

for i in names:
    for ii in mood:
        print(i, "is", ii)
```

```python
# Write a program that takes a number from the user, assigns it to the variable (number),
# and prints the squares of all numbers between 0 and the number (not included) entered by the user.
# Note: Write the program with while statement.

number = int(input('Please enter a number: '))
counter = 0
while counter < number:
    print(counter**2)
    counter += 1
```

```python
# Write a program that prints each item and its corresponding type from the following list.
# sample_list = [{"section":5, "topic":2}, 'clarusway', [1, 4], 2020, 3.14, 1+618j, False, (10, 20)]
# One of the output line as an example: The type of clarusway is <class 'str'>.
# Note: Write the program with for statement

sample_list = [{"section":5, "topic":2}, 'clarusway', [1, 4], 2020, 3.14, 1+618j, False, (10, 20)]
for i in sample_list :
    print('The type of {} is'.format(i), type(i))
```
