```python
try_tuple = ('love') # iki tırnak arasındaki değer stringdir.
                     # 'love' dan sadece virgül koyup bıraksaydık bu bir tuple olurdu. 
print(try_tuple)

print(type(try_tuple)) # it's not tuple type.
```

```python
try_tuple = ('love',) # virgülden dolayı

print(try_tuple)

print(type(try_tuple)) # it's a tuple type.
```

```python
# TUPLE FONSİYONU İÇİNE İTERABLE ALIR
# ELEMANLARINA AYIRIR
# VE ONUN  ELEMANLARINI TUPLE ELEMANI HALİNE GETİRİR.
```

```python
my_tup = ("Romania", "Poland", "Estonia")
(ro, po, es) = my_tup # elaman sayıları aynı olmalı, tek tek atama yapar python
print(ro + ", " + po + ", " + es) #returns 'Romania, Poland, Estonia'
```

```python
my_tup = ("Romania", "Poland", "Estonia")
ro = my_tup[0] # tek tek indexe göre atama yapılabilir
po = my_tup[1]
es = my_tup[2]
print(ro + ", " + po + ", " + es)
```

```python
# max ve min kullanımı
my_tup = ("Romania", "Poland", "Estonia", "Bulgaria", "Slovakia", "Slovenia", "Hungary") 
last = max(my_tup)   # en büyük değeri sıralama yapmadan verir
# last = sorted(my_tup)[-1] sıralama yapar ve en son indexi çağırır

first = min(my_tup)  # en küçük değeri sıralama yapmadan verir
# first = sorted(my_tup)[0] sıralama yapar ve ilk indexi çağırır
print(last, first)
```

```python
# Liste değiştirilebilir, (mutable) onun için hata vermez.

city_list = ['Tokyo', 'Istanbul', 'Moskow', 'Dublin']

city_list[0] = 'Athens' # Tokyo'nun yerini aldı
city_list[1] = 'Cairo' # İstanbul yerini alır.
print(city_list)
```

```python
# TUPLE değiştirilemez, (immutable) onun için hata verir..

city_list = ['Tokyo', 'Istanbul', 'Moskow', 'Dublin']

city_tuple = tuple(city_list)

city_tuple[0] = 'New York'  # you can't assign a value, hata verir.
```
