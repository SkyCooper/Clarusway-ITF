```python
def parrot(voltage, state='a stiff', action='voom', type='Norwegian Blue'):
    print("-- This parrot wouldn't", action, end=' ')
    print("if you put", voltage, "volts through it.")
    print("-- Lovely plumage, the", type)
    print("-- It's", state, "!")
```

```python
parrot() # içinde 1 positional argüman olması lazım
```

```python
parrot(1000)
```

```python
parrot(voltage=1000000, action = "VOOOOOOMMM") #kwargs ile çağırma
```

```python
parrot('a thousand', state='pushing up the daisies')  # 1 positional--> ilk değere gider /voltage , 1 keyword
```

```python
parrot(1000)                                          # 1 positional argument
parrot(voltage=1000)                                  # 1 keyword argument
parrot(voltage=1000000, action='VOOOOOM')             # 2 keyword arguments
parrot(action='VOOOOOM', voltage=1000000)             # 2 keyword arguments 
parrot('a million', 'bereft of life', 'jump')         # 3 positional arguments
parrot('a thousand', state='pushing up the daisies')  # 1 positional, 1 keyword
```

# *args

```python
def fruiterer(*fruit) : # buraya dah sonradan istediğim sayıda parametre ekleyebilirim 1.....21...101 vs. sayı belli değil
    print('I want to get :')  #*fruit artık iterable oldu. for döngüsünde kullanılabilir.
    for i in fruit :
        print('-', i)
        
fruiterer('orange', 'banana', 'melon', 'ananas') 
```

```python
def fruiterer(*fruit) :
    print(type(fruit)) # tuple oldu
fruiterer('orange', 'banana', 'melon', 'ananas')
```

```python
odd_list = []
even_list = []
def slicer(*number):
    for i in number:
        if i%2==0:
            even_list.append(i)
        else:
            odd_list.append(i)
    print("even list :", even_list)
    print("odd list :", odd_list)
slicer(1,2,3,4,5,6,7,8,9)
```

```python
def slicer(*number):
    evens = [number[1::2]]   #slicelema ile yapıldığından neyi slice edersen sonuç o olur
    odds = [number[::2]]     #tuple slice ettiğimiz için çıktı tuple oldu.
    print("even list :", evens)
    print("odd list :", odds)
slicer(1,2,3,4,5,6,7,8,9)
```

# **kwargs

```python
def animals(**kwargs):
    for key, value in kwargs.items(): # dictionary gibi davranıyoruz. dicy metodlarını kullanabilir item--> key/value den oluşur
        print(value, "are", key)
 
animals(Carnivores="Lions", Omnivores="Bears", Herbivores="Deers", Nomnivores="Human")
```

```python
def organizer(** people):
    names = []
    ages = []
    for key, value in people.items():
        names.append(key)
        ages.append(value)
        
    print(names, ages, sep="\n")
 
organizer(Beth=26, Oscar=42, Justin=18, Frank=33)
```

```python
# aynı foksiyonu List Comprehansion ile yapma;

def organizer(** data):
    names = [i for i in data.keys()] # her iterasyonda sırayla bir key ekle demek..
    ages = [ i for i in data.values()] # sırayla bir value ekle demek.
    
    print(names, ages, sep="\n")
    
organizer(Asuman=24, Oscar=42, Justin=18, Frank=33)
```

```python
def gene(x, y):  # defined by positional args
    print(x, "belongs to Generation X")
    print(y, "belongs to Generation Y")
 
dict_gene = {'y' : "Marry", 'x' : "Fred"}
gene(**dict_gene)  # we call the function by a single argument(variable)
```

```python
def gene(x='Solomon', y='David'):  # defined by kwargs (default values assigned to x and y)
    print(x, "belongs to Generation X")
    print(y, "belongs to Generation Y")
 
dict_gene = {'y' : "Marry", 'x' : "Fred"}
gene(**dict_gene) 
```
